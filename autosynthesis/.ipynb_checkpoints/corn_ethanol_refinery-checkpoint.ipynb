```python
# Utils
def get_ethanol_purity(): # wt% as decimal
    return ethanol.imass['Ethanol']/ethanol.F_mass

def get_IRR_for_given_ethanol_price(ethanol_price=0.51078): # $/kg-ethanol 
    ethanol.price = ethanol_price * get_ethanol_purity()
    return corn_tea.solve_IRR() # percent as decimal

def get_minimum_ethanol_selling_price_for_given_IRR(IRR=0.10): # percent as decimal
    corn_tea.IRR = IRR
    return corn_tea.solve_price(ethanol)/get_ethanol_purity() # $/kg-ethanol 

def set_feedstock_capacity(feedstock_capacity): # wet metric ton/y
    feedstock.F_mass = feedstock_capacity * 1e3 / corn_tea.operating_hours
    corn_sys.simulate()
    
def get_feedstock_capacity():
    return feedstock.F_mass * corn_tea.operating_hours / 1e3 # wet metric ton/y

def get_ethanol_production_capacity():
    return ethanol.F_mass * get_ethanol_purity() * corn_tea.operating_hours / 1e3 # metric ton-ethanol/y

def plot_IRR_across_feedstock_capacities(ethanol_price=0.51078, # $/kg-ethanol 
                                         feedstock_capacities=np.linspace(160000, 500000, 25), # wet metric ton/y
                                         ):
    IRRs = []
    for i in feedstock_capacities:
        set_feedstock_capacity(i)
        IRRs.append(get_IRR_for_given_ethanol_price(ethanol_price))
    
    plt.plot(feedstock_capacities, IRRs)
    return IRRs
```

```python
# Plots
corn_sys.diagram('cluster')
plot_IRR_across_feedstock_capacities(ethanol_price=0.51078, # $/kg-ethanol 
                                     feedstock_capacities=np.linspace(160000, 500000, 25), # wet metric ton/y
                                     )
```

```python
# Quick single-point results

set_feedstock_capacity(300_000) # wet metric ton/y
print('Total capital investment (MM$): ', corn_tea.TCI/1e6)
print('Annual operating cost (MM$/y): ', corn_tea.AOC/1e6)
print('Annual corn cost (MM$/y) :', feedstock.cost*corn_tea.operating_hours/1e6)
print('Annual ethanol sales (MM$/y): ', ethanol.cost*corn_tea.operating_hours/1e6)
print('\n')
print(get_IRR_for_given_ethanol_price(ethanol_price=0.55)) # $/kg-ethanol; percent as decimal
print(get_minimum_ethanol_selling_price_for_given_IRR(IRR=0.10)) # percent as decimal; $/kg-ethanol
```
